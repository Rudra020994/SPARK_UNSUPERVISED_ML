```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

# Importing dataset

```python
df= pd.read_csv("D:\\Iris.csv")
```

```python
df
```

# Considering petal length and width to obtain the clusters

```python
x = df[["PetalLengthCm","PetalWidthCm"]]
```

```python
x
```

# Doing Elbow plot to get the optiman number of clusters

```python
from sklearn.cluster import KMeans
wcss = []
for i in range(1,11):
    Kmeans = KMeans(n_clusters = i)
    Kmeans.fit(x)
    wcss.append(Kmeans.inertia_)
```

```python
plt.plot(range(1,11),wcss)
```

# Considering 3 as the optimal number of clusters

```python
Km = KMeans(n_clusters = 3)
```

```python
Km
```

```python
y_kmeans = Km.fit_predict(x)
```

# The entire dataset have been divided into 3 clusters (0,1,2)

```python
y_kmeans
```

```python
df['clusters'] = y_kmeans
```

```python
df
```

# Visualising the result 

```python
m = df[df.clusters == 0]
n = df[df.clusters == 1]
p = df[df.clusters == 2]
```

```python
m
```

# Initial plot without clustering

```python
plt.scatter(df.PetalLengthCm,df.PetalWidthCm)
```

# Plotting after clustering

```python
plt.scatter(m['PetalLengthCm'],m['PetalWidthCm'],color = 'blue')
plt.scatter(n['PetalLengthCm'],n['PetalWidthCm'],color = 'green')
plt.scatter(p['PetalLengthCm'],p['PetalWidthCm'],color = 'black')
```

# Calculating cluster centres

```python
Km.cluster_centers_
```

# Plotting with cluster centers

```python
plt.scatter(m['PetalLengthCm'],m['PetalWidthCm'],color = 'blue')
plt.scatter(n['PetalLengthCm'],n['PetalWidthCm'],color = 'green')
plt.scatter(p['PetalLengthCm'],p['PetalWidthCm'],color = 'black')
plt.scatter(Km.cluster_centers_[:,0],Km.cluster_centers_[:,1], color = 'yellow')
```

# Trying to clusted the dataset using sepal length and width

# Scaling the data

```python
from sklearn.preprocessing import MinMaxScaler
```

```python
scaler = MinMaxScaler()
scaler.fit(df[['SepalLengthCm']])
df['SepalLengthCm'] = scaler.transform(df[['SepalLengthCm']])
```

```python
scaler = MinMaxScaler()
scaler.fit(df[['SepalWidthCm']])
df['SepalWidthCm'] = scaler.transform(df[['SepalWidthCm']])
```

```python
x1 = df[["SepalLengthCm","SepalWidthCm"]]
```

```python
x1
```

# Obtaining the Elbow plot

```python
from sklearn.cluster import KMeans
wcss = []
for i in range(1,11):
    Kmeans = KMeans(n_clusters = i)
    Kmeans.fit(x1)
    wcss.append(Kmeans.inertia_)
```

```python
plt.plot(range(1,11),wcss)
```

# Optimal number of clusters is found o be 5

```python
Km = KMeans(n_clusters = 5)
y_kmeans = Km.fit_predict(x1)
df['clusters'] = y_kmeans
```

```python
df
```

```python
m = df[df.clusters == 0]
n = df[df.clusters == 1]
p = df[df.clusters == 2]
q = df[df.clusters == 3]
r = df[df.clusters == 4]
```

```python
m
```

```python
Km.cluster_centers_
```

# Visualising the plot with cluster centers

```python
plt.scatter(m['SepalLengthCm'],m['SepalWidthCm'],color = 'blue')
plt.scatter(n['SepalLengthCm'],n['SepalWidthCm'],color = 'green')
plt.scatter(p['SepalLengthCm'],p['SepalWidthCm'],color = 'black')
plt.scatter(q['SepalLengthCm'],q['SepalWidthCm'],color = 'red')
plt.scatter(r['SepalLengthCm'],r['SepalWidthCm'],color = 'pink')
plt.scatter((Km.cluster_centers_[:,0]),(Km.cluster_centers_[:,1]), color = 'yellow')
```
